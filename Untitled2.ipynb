Ejercicio

Se tienen los puntos A y B en el cuadrante positivo del plano cartesiano, elabore el programa que permita obtener la distancia entre A y B.

D: √(Ax-Bx)^2+(Ay-By)^2

```python
print("---------------------------------------------------")
print("-------Calcular la distancia entre dos puntos------")
print("---------------------------------------------------")

#Datos de entrada
print("Calcular la distancia entre dos puntos  ")
Ax=int(input("Ax: "))
Ay=int(input("Ay: "))
Bx=int(input("Bx: "))
By=int(input("By: "))

#Proceso
D=((Ax-Bx**2)+((Ay-By)**2)**0.5)

#Salida
print("La distancia entre los puntos es: ",D)
```
